```python
import pandas as pd
import numpy as np

df=pd.read_excel('Superstore -1.xlsx')

# x=[24,82,75,19,10,21,23,44,23,21,49]
# y=[1,4,2,5,9,8,2,1,2,4,4]
class SSS:
#     import numpy as np
    def __init__(self,mylist,mylist2=None):
        self.mylist=mylist
        self.mylist2=mylist2
       
   
    def summ(self):
        import numpy as np
        sum_of_x=np.sum(self.mylist)
        print(f'sum_of_x = {sum_of_x}')
        
 
    def Length(self):
        len_of_x=len(self.mylist)
        print(f'len_of_x = {len_of_x}')

      
    def avg(self):
        import numpy as np
        mean_of_x=round(np.mean(self.mylist),2)
        print(f'mean_of_x = {mean_of_x}')
        
  
    def median(self):
        '''This function calculates median of the given list.
        Example:-
            x_object=(SSS([1,2,2,3,34,5,5,7,8,8,9,4]))
            x_object.median()'''
        import numpy as np
        median_of_x=(np.median(self.mylist))
        print(f'median_of_x = {median_of_x}')
        

    def mode(self):
        total_count={}
        for i in self.mylist:
            try:
                if total_count[i]>=1:
                    total_count[i]=total_count[i]+1
                else:
                    pass
            except:
                total_count[i]=1

        mode_val=[]
        for i,j in total_count.items():
            if j == max(total_count.values()):
                mode_val.append(i)
            else:
                pass
        print(sorted(f'mode_val = {mode_val}'))
    
    def variance(self):
        import numpy as np
        x_var=round(np.var(self.mylist),2)
        print(f'x_var = {x_var}')
      
    #7) standard deviation
    def standard_deviation(self):
        import numpy as np
        x_var=round(np.var(self.mylist),2)
        x_std=round(np.sqrt(x_var),2)
        print(f'x_std = {x_std}')
    
    #8) covariance
    def covariance(self,mylist2):
        import numpy as np
        import statistics as sts
        cov_xy=round(sts.covariance(self.mylist,self.mylist2),2)
        print(f'cov_xy = {cov_xy}')
        
    #9) correlation
    def correlation(self):
        import numpy as np
        import statistics  as sts
        corr_xy=round(sts.correlation(self.mylist,self.mylist2),2)
        print(f'corr_xy = {corr_xy}')
  
    #10) range
    def Range(self):
        import numpy as np
        x_range=np.ptp(self.mylist)
        print(f'x_range = {x_range}')
       
    #11) Q1
    def perc_Q1(self):
        import numpy as np
        q1_x=np.percentile(self.mylist,25)
        print(f'q1_x = {q1_x}')
        

    #12) Q3
    def perc_q3(self):
        import numpy as np
        q3_x=np.percentile(self.mylist,75)
        print(f'q3_x = {q3_x}')
       
    #13) IQR
    def Interquartile_range(self):
        import numpy as np
        q1_x=np.percentile(self.mylist,25)
        q3_x=np.percentile(self.mylist,75)
        iqr_x=q3_x-q1_x
        print(f'iqr_x = {iqr_x}')
        

    #14) Confidence Interval 95%,97%,99%
    def Confidence_interval(self):
        import numpy as np
        import statistics as sts
        mean_of_x=np.mean(self.mylist)
        len_of_x=len(self.mylist)
        x_std=sts.stdev(self.mylist)
        
        ci_x_95=round(mean_of_x-(1.96*(x_std/(np.sqrt(len_of_x)))),2),round(mean_of_x+(1.96*(x_std/(np.sqrt(len_of_x)))),2)
        print(f'ci_x_95 = {ci_x_95}')

    def all_stats(self):
        self.summ()
        self.Length()
        self.avg()
        self.median()
        self.mode()
        self.variance()
        self.standard_deviation()
        self.Range()
        self.perc_Q1()
        self.perc_q3()
        self.Interquartile_range()
        self.Confidence_interval()

# In[ ]:





# In[3]:

if __name__=='__main__':
    
    print('Profit:')
    superstore_profit=SSS(df['profit'])
    superstore_profit.all_stats()
    print('*'*70)
    
    print('Sales:')
    superstore_sales=SSS(df['sales'])
    superstore_sales.all_stats()
    print('*'*70)
    
    print('Discount:')
    superstore_discount=SSS(df['discount'])
    superstore_discount.all_stats()
    print('*'*70)
    
    print('Quantity:')
    superstore_quantity=SSS(df['quantity'])
    superstore_quantity.all_stats()
    print('*'*70)
    
   
    
    
```
